```python
import time
from concurrent.futures import ThreadPoolExecutor, as_completed
```

### Concurrencia

Utilizar o no el metodo as_completed() puede resultar entre que una tarea avance mientras espera respuesta o se quede estancada esperando un elemento.


```python
def cooking_fast_task(name, seconds):
    time.sleep(seconds)
    return f"{name} termino en {seconds}s"

def cooking_slow_task(name, seconds):
    time.sleep(seconds)
    return f"{name} termino en {seconds}s"
```

```python
def basic_fast_example():
    tasks = [
        ("Hornear Pavo", 5),   # Tarda mucho
        ("Servir Refresco", 1), # Tarda poco
        ("Poner Mesa", 2)       # Tarda medio
    ]

    with ThreadPoolExecutor(max_workers=3) as executor:
        future_to_dish = {executor.submit(cooking_fast_task, t[0],t[1]): t[0] for t in tasks}

        print("---Con as_completed")
        for future in as_completed(future_to_dish):
            result = future.result()
            print(result)
```

```python
def basic_slow_example():
    tasks = [
        ("Hornear Pavo", 5),   # Tarda mucho
        ("Servir Refresco", 1), # Tarda poco
        ("Poner Mesa", 2)       # Tarda medio
    ]

    with ThreadPoolExecutor(max_workers=3) as executor:
        future_to_dish = {executor.submit(cooking_slow_task, t[0],t[1]): t[0] for t in tasks}

        print("\n---Sin as_completed")
        for future in future_to_dish:
            result = future.result()
            print(result)
```

```python
basic_fast_example()
basic_slow_example()
```
